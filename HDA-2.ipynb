```python
df=pd.DataFrame(data)
#  select columns that are not 'Gene ID'
value_columns = df.columns.difference(['Gene ID', 'QCPool1', 'QCPool2', 'QCPool3'])

# Initialize an empty dictionary to store counts for different cutoffs
count_dict = {}

# Iterate over different cutoff values and store counts
for cutoff in range(1, len(value_columns) + 1):
    non_nan_count_TC = df[[col for col in value_columns if "TC" in col]].count(axis=1)
    non_nan_count_NC = df[[col for col in value_columns if "NC" in col]].count(axis=1)
    rows_to_keep = (non_nan_count_TC >= cutoff) | (non_nan_count_NC >= cutoff)
    count_dict[cutoff] = sum(rows_to_keep)

# Create a DataFrame from the count dictionary
count_table = pd.DataFrame(count_dict.items(), columns=['TC and NC Cutoff', 'Protein Count'])

# Display the count table
print(count_table)
```

# Data Cleaning

```python
df = pd.DataFrame(data)
df.set_index('Gene ID', inplace=True)
```

```python
# tc_columns = [col for col in df.columns if 'TC' in col]
# nc_columns = [col for col in df.columns if 'NC' in col]

# tc_data = df[tc_columns]
# nc_data = df[nc_columns]
```

```python
# Count the number of non-NaN values for each row in TC and NC columns
non_nan_count_TC = df[[col for col in value_columns if "TC" in col]].count(axis=1)

# Identify rows with at least 6 non-NaN entries in TC columns
rows_to_keep = (non_nan_count_TC >= 6)

# Filter rows
tc_retain_df = df[rows_to_keep]
tc_columns = [col for col in tc_retain_df.columns if 'TC' in col]
tc_data = tc_retain_df[tc_columns]
tc_data
```

```python
# Count the number of non-NaN values for each row in TC and NC columns
non_nan_count_NC = df[[col for col in value_columns if "NC" in col]].count(axis=1)

# Identify rows with at least 5 non-NaN entries in NC columns
rows_to_keep = (non_nan_count_NC >= 6)

# Filter rows
nc_retain_df = df[rows_to_keep]
nc_columns = [col for col in df.columns if 'NC' in col]
nc_data = nc_retain_df[nc_columns]
nc_data
```

```python
# Initialize KNNImputer with the desired number of neighbors
knn_imputer = KNNImputer(n_neighbors=5)

# Convert the DataFrame to a NumPy array for imputation
data_tc_impute = tc_data.values

# Perform imputation
imputed_tc_data = knn_imputer.fit_transform(data_tc_impute)

# Convert the imputed values back to a DataFrame
imputed_tc_df = pd.DataFrame(imputed_tc_data, columns=tc_data.columns, index=tc_data.index)

# Display the final imputed DataFrame
imputed_tc_df
```

```python
# Initialize KNNImputer with the desired number of neighbors
knn_imputer = KNNImputer(n_neighbors=1)

# Convert the DataFrame to a NumPy array for imputation
data_nc_impute = nc_data.values

# Perform imputation
imputed_nc_data = knn_imputer.fit_transform(data_nc_impute)

# Convert the imputed values back to a DataFrame
imputed_nc_df = pd.DataFrame(imputed_nc_data, columns=nc_data.columns, index=nc_data.index)

# Display the final imputed DataFrame
imputed_nc_df
```

```python
common_data = imputed_nc_df.merge(imputed_tc_df, left_index=True, right_index=True)
common_data = common_data.T
common_data
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
# Assuming you have a DataFrame df and it was previously Min-Max scaled
df = pd.read_csv('tumor_impute.csv')
```

```python
df.shape
```

```python
scaler = MinMaxScaler()
normalized_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
normalized_df
```

```python
# Define the minimum and maximum values used in scaling
min_values = df.min()
max_values = df.max()

# Denormalize the DataFrame
denormalized_df = normalized_df

for column in df.columns:
    denormalized_df[column] = normalized_df[column] * (max_values[column] - min_values[column]) + min_values[column]
```

```python
denormalized_df=pd.DataFrame(denormalized_df, columns=denormalized_df.columns, index = denormalized_df.index)
```
